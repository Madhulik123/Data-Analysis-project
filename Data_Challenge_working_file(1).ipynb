```python
# prompt: find all the symbol or character in the dataset which should not be part of the column, find for all columns

import string

# Function to find non-standard characters in a column
def find_non_standard_chars(df, column):
    # Define allowed characters (letters, numbers, whitespace, common punctuation)
    allowed_chars = string.ascii_letters + string.digits + string.whitespace + string.punctuation

    # Initialize a set to store unique non-standard characters
    non_standard_chars = set()

    # Iterate through each value in the column
    for value in df[column].dropna(): # Use dropna to skip NaN values
        # Convert value to string (handle potential non-string data types)
        value_str = str(value)
        # Check each character in the string
        for char in value_str:
            if char not in allowed_chars:
                non_standard_chars.add(char)
    return non_standard_chars

# Iterate through all columns in the DataFrame
for col in df.columns:
    non_standard_chars_in_col = find_non_standard_chars(df, col)
    if non_standard_chars_in_col:
        print(f"Non-standard characters found in column '{col}': {non_standard_chars_in_col}")
    else:
        print(f"No non-standard characters found in column '{col}' (or column is empty/contains only standard chars).")
```

#Is there a correlation between the language used for search and the user's location or destination?

```python
import pandas as pd
from scipy.stats import chi2_contingency
# Create a contingency table of 'website_language' and 'user_country_source1'
contingency_table = pd.crosstab(df['website_language'], df['user_country_source1'])

# Perform chi-square test for independence
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print("Chi-square test results for correlation between website language and user's country:")
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")

# Create a contingency table of 'website_language' and 'searched_destination'
contingency_table_dest = pd.crosstab(df['website_language'], df['searched_destination'])

# Perform chi-square test for independence
chi2_dest, p_dest, _, _ = chi2_contingency(contingency_table_dest)

# Print the results
print("\nChi-square test results for correlation between website language and searched destination:")
print(f"Chi-square statistic: {chi2_dest}")
print(f"P-value: {p_dest}")
```

```python
# #
# The chi-square test results indicate that there is a statistically significant correlation between the website language and both the user's country and the searched destination.

# With extremely low p-values (p < 0.05) and large chi-square statistics, we reject the null hypothesis, suggesting that there is a significant association between the website language and the user's country as well as the searched destination.

# This suggests that the choice of website language is not independent of the user's country or the destination they are searching for, indicating a strong correlation between these variables.
```

What is the average lead time for users between performing a search and their planned departure date?

```python
import pandas as pd

# Assuming df is your DataFrame containing the dataset

# Convert 'search_ts' column to datetime
df['search_ts'] = pd.to_datetime(df['search_ts'])

# Calculate lead time by adding 'days_to_departure' to 'search_ts'
df['planned_departure_date'] = df['search_ts'] + pd.to_timedelta(df['days_to_departure'], unit='D')

# Calculate lead time by subtracting 'search_ts' from 'planned_departure_date'
df['lead_time'] = df['planned_departure_date'] - df['search_ts']

# Convert lead time to days
df['lead_time_days'] = df['lead_time'] / pd.Timedelta(days=1)

# Calculate average lead time
average_lead_time = df['lead_time_days'].mean()

print("Average lead time for users between performing a search and their planned departure date: {:.2f} days".format(average_lead_time))
```

Is there a correlation between the distance of the searched destination from the user's city and the lead time?

```python

# Check for correlation between 'distance' and 'lead_time_days'
correlation = df['distance'].corr(df['lead_time_days'])

print("Correlation between distance of the searched destination and lead time: {:.2f}".format(correlation))
plt.figure(figsize=(8, 6))
plt.scatter(df['distance'], df['lead_time_days'], alpha=0.5)
plt.title('Correlation between Distance of Searched Destination and Lead Time')
plt.xlabel('Distance')
plt.ylabel('Lead Time (days)')
plt.grid(True)
plt.show()

```
